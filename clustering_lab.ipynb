```python
import numpy as np
import pandas as pd
```

```python
products = pd.read_csv('product.csv', names=['name', 'degree'])
```

```python
products.head()
```

```python
p_name = products['name']
p_name.head()
```

```python
p_degree = products['degree'][:30]
p_degree
```

```python
orderCount = np.random.randint(1, 15) # 訂單數
order = pd.DataFrame(columns=['no', 'name', 'degree', 'count'])
for i in range(2):
    
    ol_data = pd.DataFrame(columns=['no', 'name', 'degree', 'count'])
    ol_count = np.random.randint(1, 30)
    
    
    ol_data['name'] = np.random.choice(p_name, ol_count, replace=False) # 出貨單品項
    ol_data['degree'] = np.random.choice(p_degree, ol_count)
    ol_data['count'] = np.random.randint(1, 20, ol_count)
    ol_data['no'] = i + 1

    if i == 0:
        order = ol_data
    else:
        order = pd.concat([order, ol_data], ignore_index=True)
        
print(order)

        
```
